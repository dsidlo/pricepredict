# All Trading Pairs - Data Analysis



```python
import pandas as pd

# Load Data/all_trading_pairs.csv

all_trading_pairs = pd.read_csv('Data/all_trade_pairs.csv')

# Find Count trading pairs that have more than 1 record
# resulting in a dataframe with the trading pairs and the count of records
trading_pairs_count = all_trading_pairs.groupby('potential_pair').size().reset_index(name='counts')
# Sort the dataframe by the counts column
trading_pairs_count = trading_pairs_count.sort_values(by='counts', ascending=False)

trading_pairs_count
```

```python
# Given all_trading_pairs, find trading pairs that have more than 1 record and that have a period_days of
# less than or equal to 20.
trading_pairs_count = all_trading_pairs[all_trading_pairs['period_days'] <= 60].groupby('potential_pair').size().reset_index(name='counts')

trading_pairs_count
```
