```python

'''
The following can be used to get 5min bars from the yahoo finance API.
This data can be used to predict a 5min and/or 15min price movement.
  
'''
import pandas as pd

import yfinance as yf
import datetime
import pytz


def get_price_bars(symbol) -> pd.DataFrame:
    # Get data for today
    eastern = pytz.timezone('US/Eastern')
    now = datetime.datetime.now(eastern)
    start = now.strftime('%Y-%m-%d')
    end = now + datetime.timedelta(days=1)
    end = end.strftime('%Y-%m-%d')

    # Fetch data
    # Interval can be '1m', '5m', '15m', '30m', '1h', '1d'
    # data = yf.download(symbol, start=start, end=end, interval='5m')
    # data = yf.download(symbol, start=start, end=end, interval='1h')
    data = yf.download(symbol, start=start, end=end, interval='1m')
    
    return data


bars = get_price_bars('^IXIC') 

bars
```
