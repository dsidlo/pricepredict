```python
# Read a test file of S&P 500 stock symbols.
# Validate them against yfinance Tickers function.

import yfinance as yf

# Read the file "SP500-25-01-19.txt". Its format is: <Symbol>: <Company Name>
symbols = []
with open('SP500-25-01-19.txt', 'r') as f:
    lines = f.readlines()
    for l in lines:
        symbol = l.split(':')[0].strip()
        symbols.append(symbol)
        
tickers = yf.Tickers(" ".join(symbols))

valid_symbols = {}
invalid_symbols = {}
for sym in symbols:
    try:
        ticker_info = tickers.tickers[sym].info
        if ticker_info is not None:
            valid_symbols[sym] = ticker_info
    except Exception as e:
        print(f"*** Exception for symbol: {sym}\n->|| {e} ||")    
        invalid_symbols[sym] = e

print("Valid Symbols: ", valid_symbols.keys())
print("Invalid Symbols: ", invalid_symbols.keys())

pass
```
