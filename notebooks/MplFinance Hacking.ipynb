## How to create overlapping plots with the main plit hidden.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch historical data for gold and IOVA
gold_df = yf.download('GC=F', start='2020-01-01', end='2023-08-01')
iova_df = yf.download('IOVA', start='2020-01-01', end='2023-08-01')

# Align data on dates (common timestamps)
data = pd.DataFrame({
    'Gold (GC=F)': gold_df['Close'],
    'IOVA': iova_df['Close']
}).dropna()
fig, ax1 = plt.subplots()
# ax2 = ax1.twinx()
# ax1 = plt.twinx()
ax2 = ax1.twinx()

# Plotting the data
plt.figure(figsize=(12, 6))

ax1.set_title('Comparison: Gold (GC=F) vs IOVA Stock Prices')
ax1.set_ylabel('Gold (GC=F)')
ax1.plot(data.index, data['Gold (GC=F)'], label='Gold (GC=F)', color='gold')
ax2.set_ylabel('IOVA')
ax2.plot(data.index, data['IOVA'], label='IOVA', color='blue')
fig.tight_layout()

# plt.title('Comparison: Gold (GC=F) vs IOVA Stock Prices')
# plt.legend()
# plt.grid(True)

plt.show()
```
